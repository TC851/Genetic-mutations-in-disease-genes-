The project I am conducting focuses on analysing exome or genome sequencing data with the aim of identifying genetic mutations in disease genes. This is particularly important in order to gain a better understanding of the genetic basis of diseases and potentially develop new approaches for diagnoses and therapies.

The main objective of this project is to identify genetic mutations in disease genes by basic analysis of exome or genome sequencing data.

```python
import os
import gzip

directory = "/Users/caner/Desktop/Humangenetik/Mutationen"

url = "ftp://ftp.1000genomes.ebi.ac.uk/vol1/ftp/phase3/data/HG00096/sequence_read/SRR062634.filt.fastq.gz"
output_file = os.path.join(directory, "SRR062634.filt.fastq.gz")

os.system(f"wget {url} -O {output_file}")
```

```python
num_lines = 10

with gzip.open(output_file, 'rt') as file:
    for _ in range(num_lines):
        line = file.readline()
        print(line.strip())
```
