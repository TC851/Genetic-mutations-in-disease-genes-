```python
import os
import gzip
from IPython.display import Image, display, HTML
```

```python
directory = "/Users/caner/Desktop/Humangenetik/Mutationen"
os.makedirs(directory, exist_ok=True)

urls = [
    "ftp://ftp.sra.ebi.ac.uk/vol1/fastq/SRR605/SRR605663/SRR605663_1.fastq.gz",
    "ftp://ftp.sra.ebi.ac.uk/vol1/fastq/SRR605/SRR605663/SRR605663_2.fastq.gz"
]
for url in urls:
    output_file = os.path.join(directory, os.path.basename(url))
    os.system(f"wget {url} -O {output_file}")
```

```python
file1 = "/Users/caner/Desktop/Humangenetik/Mutationen/SRR605663_1.fastq.gz"
num_lines = 12

first_lines_file1 = []
with gzip.open(file1, 'rt') as file:
    for _ in range(num_lines):
        line = file.readline().strip()
        first_lines_file1.append(line)
        
file2 = "/Users/caner/Desktop/Humangenetik/Mutationen/SRR605663_2.fastq.gz"

first_lines_file2 = []
with gzip.open(file2, 'rt') as file:
    for _ in range(num_lines):
        line = file.readline().strip()
        first_lines_file2.append(line)
         
first_lines_file1, first_lines_file2
```

```python
fastqc_reports_dir = os.path.join(directory, "FastQC_Reports")
os.makedirs(fastqc_reports_dir, exist_ok=True)

fastq_files = [
    "/Users/caner/Desktop/Humangenetik/Mutationen/SRR605663_1.fastq.gz",
    "/Users/caner/Desktop/Humangenetik/Mutationen/SRR605663_2.fastq.gz"
]
```

```python
for fastq_file in fastq_files:
    os.system(f"fastqc -o {fastqc_reports_dir} {fastq_file}")

os.system(f"multiqc {fastqc_reports_dir} -o {fastqc_reports_dir}")

multiqc_report = os.path.join(fastqc_reports_dir, "multiqc_report.html")
if os.path.exists(multiqc_report):
    multiqc_url = f"file://{multiqc_report}"
    display(HTML(f"<a href='{multiqc_url}' target='_blank'>MultiQC Report anzeigen</a>"))
else:
    display(HTML("<p>MultiQC Report not found at expected location</p>"))

created_files = os.listdir(fastqc_reports_dir)
created_files
```
